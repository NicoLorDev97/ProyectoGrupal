```python
# tbl_srv_sanit_bas
"""
VEN, ARG: Hay valores faltantes en inicio de perÃ­odo, los valores van en crecimiento
se aplicarÃ¡ interpolaciÃ³n o bfill

BHS: Solo hay un valor faltante. Se aplicarÃ¡ bfill, ya que se ve que los valores
de los aÃ±os cercanos son iguales

VCT, KNA, GRD, DMA, ATG: Caso similar a BHS. Se aplicarÃ¡ bfill
"""
tbl_srv_sanit_bas = pd.read_csv('Extraccion/tbl_srv_sanit_bas.csv')
```

```python
for pais in paises:
    temp = tbl_srv_sanit_bas[tbl_srv_sanit_bas['id_pais']==pais]
    if pais == 'VEN':
        temp = temp.interpolate()
        tbl_srv_sanit_bas[tbl_srv_sanit_bas['id_pais']==pais] = temp
    else:
        temp = tbl_srv_sanit_bas[tbl_srv_sanit_bas['id_pais']==pais]
        temp = temp.fillna(method = 'bfill')
        tbl_srv_sanit_bas[tbl_srv_sanit_bas['id_pais']==pais] = temp
tbl_srv_sanit_bas.to_csv('revisar.csv')
```

```python
print("Paises con TODOS sus datos nulos")
paises_nulos=[]
for i in paises:
    mask = alfa["id_pais"] == i
    df_filtrada = alfa[mask]
    a = df_filtrada.iloc[:,2].isnull().sum()
    if a == 21:
        paises_nulos.append(i)
print(paises_nulos)
```

```python
"""
Imputaremos un valor para cada uno de ellos, para luego predecir el resto de la tabla<br>
Fuente: https://worldpopulationreview.com/country-rankings/literacy-rate-by-country<br>

CAN = Canada, AÃ±o 2003, Ind. Alfabetizacion: 99.00% <br>
DMA = Dominica, AÃ±o 2003, Ind. Alfabetizacion: 94.00%<br>
KNA = Saint Kitts and Nevis, AÃ±o 2003, Ind. Alfabetizacion: 97.80%<br>
LCA = Santa Lucia, AÃ±o 2003, Ind. Alfabetizacion: 90.10%<br>
VCT = Saint Vincent and the Grenadines, AÃ±o: 2003, Ind. Alfabetizacion: 93.00%<br>
BHS = Bahamas, AÃ±o: 2003, Ind. Alfabetizacion: 95.60%<br>
USA = Estados Unidos, AÃ±o: 2003, Ind. Alfabetizacion: 99.00%
"""
#creamos un dataframe con los valores a cargar
imp=[99.00, 94.00, 97.80, 90.10, 93.00, 95.60, 99.00]
df_carga=pd.DataFrame(list(zip(paises_nulos, imp)), columns= ["id_pais","alfabetizacion"])
df_carga
```

```python
for i in paises_nulos:
    mask1 = (alfa["id_pais"] == i) & (alfa["year"] == "2003")
    mask2 = df_carga["id_pais"] == i
    alfa.loc[mask1, "alfabetizacion"] = df_carga["alfabetizacion"].loc[mask2].values[0]
    print (alfa[mask1])
```
