HaY 567 datos faltantes<br>
Vamos a identificarlos

```python
#Vemos que paises tienen TODOS sus datos faltantes
print("Paises con TODOS datos nulos")
paises_nulos=[]
for i in paises:
    mask = tbl_consumo_tabaco["id_pais"] == i
    df_filtrada = tbl_consumo_tabaco[mask]
    a = df_filtrada.iloc[:,2].isnull().sum()
    if a == 21:
        paises_nulos.append(i)
print(paises_nulos)
```

```python
#Vemos que paises tienen datos faltantes
print("Paises con datos nulos")
paises_nulos=[]
for i in paises:
    mask = tbl_consumo_tabaco["id_pais"] == i
    df_filtrada = tbl_consumo_tabaco[mask]
    a = df_filtrada.iloc[:,2].isnull().sum()
    if a > 0:
        paises_nulos.append(i)
print(paises_nulos)
```

```python
#vemos el resultado por pais
for i in paises:
    print(tbl_consumo_tabaco[tbl_consumo_tabaco["id_pais"] == i])
```

Tenemos TODOS los datos faltantes en algunos paises, 'ATG', 'DMA', 'GRD', 'HND', 'NIC', 'KNA', 'LCA', 'VCT', 'SUR', 'TTO', 'VEN'.<br>

INTERPOLAREMOS LOS DATOS FALTANTES en aqullos paises que tienen algunos datos faltantes

```python
#Todos los paises tienen datos nulos. Vamos a interpolar en ambas direcciones
for i in paises:
    temp = tbl_consumo_tabaco[tbl_consumo_tabaco['id_pais'] == i]
    temp = temp.interpolate(method="linear",limit_direction="both")
    tbl_consumo_tabaco[tbl_consumo_tabaco['id_pais'] == i] = temp
```

```python
#CORROBORAMOS VALORES NULOS
tbl_consumo_tabaco.isnull().sum()
```

```python
#vemos el resultado por pais
for i in paises:
    print(tbl_consumo_tabaco[tbl_consumo_tabaco["id_pais"] == i])
```
